**Iterated PageRank Vector**

The simulation was then iterated until the algorithm converged. It took 28 iterations for the PageRank values for each country to stabilize.

*The added country ranks 5th out of 8 countries (with a PageRank vector of 12.14) after the iteration.

```python
M_extended = np.array(
        [[0, 1 / 10, 1 / 7, 1 / 25, 1 / 21, 1 / 20, 0, 0],
        [0, 0, 1 / 7, 0, 0, 0, 0, 0],
        [1 / 8, 1 / 10, 0, 1 / 25, 1 / 21, 1  /20, 1 / 18, 1 / 3],
        [1 / 8, 0, 0, 0, 1 / 21, 0, 1 / 18, 1/ 3],
        [0, 0, 0, 1 / 25, 0, 1 / 20, 1 / 18, 0],
        [0, 0, 0, 1 / 25, 1 / 21, 0, 1 / 18, 1 / 3],
        [1 / 8, 1 / 10, 0, 1 / 25, 0, 1 / 20, 0, 0],
        [1 / 8, 0, 1 / 7, 0, 0, 0, 0, 0] ])

print(M_extended)
print("M_extended shape:", M_extended.shape)

R = np.ones(8) * 100 / 8
# Updated PageRank vector
print("\nUpdated PageRank vector R(0):", R)

countries = ["ZA", "GH", "NG", "RW", "UG", "KE", "ET", "AB"]
print("Countries:", countries)

# Maximum number of iterations
max_iterations = 100
# Convergence threshold
epsilon = 1e-6

t = 0
while t < max_iterations:
    R_prev = R.copy()

    # Update PageRank vector using matrix multiplication
    # (simplified rule d=1)
    R = np.dot(M_extended, R)

    # Normalize the PageRank vector to sum to 100
    R = R / np.sum(R) * 100
    print(f"\nPageRank vector R({t+1}):")
    print("\n ".join([f"{c}: {r:.2f}" for c, r in zip(countries, R)]))

    # Check for convergence
    if np.linalg.norm(R - R_prev, ord=1) < epsilon:
        print(f"\nPageRank converged after {t+1} iterations.")
        break

    t += 1
else:
    print(f"\nPageRank did not converge after {max_iterations} iterations.")
```

The PageRank Vector is more concerned with who links to you (as opposed to who you link to). The more other countries link to you, the higher your PageRank score.

Linking to a high-ranked country does not increase your PageRank entry directly. It can decrease it unless that high-ranked country (or its neighbors) link back to you, in which case the iterative process amplifies your score.

Linking to a low-ranked country does not increase your PageRank vector entry. It can decrease it indirectly, because you’re redistributing your probability into parts of the network that don’t reinforce your own score.

**Calculating Final PageRank Scores for Different Values of d**

A loop was written to calculate the final PageRank scores for all 7 countries for different values of
, ranging from 0.0 to 1.0 in steps of 0.05.


```python
import numpy as np

# The 7-country matrix M
M = np.array([
        [0, 1 / 10, 1 / 6, 1 / 25, 1 / 21, 1 / 20, 0],
        [0, 0, 1 / 6, 0, 0, 0, 0],
        [1 / 7, 1 / 10, 0, 1 / 25, 1 / 21, 1 / 20, 1 / 18],
        [1 / 7, 0, 0, 0, 1 / 21, 0, 1 / 18],
        [0, 0, 0, 1 / 25, 0, 1 / 20, 1 / 18],
        [0, 0, 0, 1 / 25, 1 / 21, 0, 1 / 18],
        [1 / 7, 1 / 10, 0, 1 / 25, 0, 1 / 20, 0], ])

countries = ["ZA", "GH", "NG", "RW", "UG", "KE", "ET"]
N = len(countries)

# damping values from 0.0 to 1.0 in steps of 0.05
d_values = np.arange(0.0, 1.05, 0.05)
max_iterations = 1000
epsilon = 1e-6

final_pageranks = {}

print("Calculating PageRank for different damping factors (d):\n")

for d in d_values:
    R = np.ones(N) * 100 / N  # Initial PageRank vector, normalized to 100
    teleportation_vector = np.ones(N) * (100 / N)

    t = 0
    while t < max_iterations:
        R_prev = R.copy()

        # PageRank update rule with damping factor d
        R = d * np.dot(M, R) + (1 - d) * teleportation_vector

        # Normalize PageRank vector to sum to 100 (if it drifts slightly)
        R = R / np.sum(R) * 100

        # Check for convergence
        if np.linalg.norm(R - R_prev, ord=1) < epsilon:
            print(f"  Converged for d={d:.2f} after {t+1} iterations.")
            break

        t += 1
    else:
        print(f"Warning: PageRank did not converge for d={d:.2f} "
              f"after {max_iterations} iterations.")

    final_pageranks[f'd={d:.2f}'] = R.copy() # Store final PageRank vector

    print(f"Final PageRank for d={d:.2f}:")
    for c, r_val in zip(countries, R):
        print(f"  {c}: {r_val:.2f}")
    print("\n" + "-"*50 + "\n")
```

**Generating a Plot**

A graph was plotted for all the 7 countries together, where the x-axis was the value of *d* and the y-axis was the PageRank score.

```python
import matplotlib.pyplot as plt

# Extract d values and PageRank scores for plotting
d_values_plot = []
pagerank_scores_by_country = {country: [] for country in countries}

for d_key, pagerank_vector in final_pageranks.items():
    d_value = float(d_key.split('=')[1])
    d_values_plot.append(d_value)
    for i, country in enumerate(countries):
        pagerank_scores_by_country[country].append(pagerank_vector[i])

# Create the plot
plt.figure(figsize=(12, 8))
for country, scores in pagerank_scores_by_country.items():
    plt.plot(d_values_plot, scores, marker='o', linestyle='-', label=country)

plt.title('PageRank Score vs. Damping Factor (d) for Each Country')
plt.xlabel('Damping Factor (d)')
plt.ylabel('PageRank Score')
plt.grid(True)
plt.legend(title='Country')
plt.xticks(d_values)
plt.tight_layout()
plt.show()
```

**Sensitivity of PageRank Scores to Changes in Damping Factor**

At *d*=0.0 (no damping), the PageRank score for all countries is equal (14.29), because the algorithm relies purely on the teleportation factor, distributing rank equally among all pages.

As *d* increases, countries with more incoming links see their PageRank score increase, and countries with less incoming links see their PageRank score decrease. It is worth noting that Rwanda's curve has a relatively stable PageRank across different *d* values suggesting that the country is benefiting both from links as well as from random jumps.

As *d* approaches 1 the PageRank score differences between the countries become much more pronounced. In essence, a higher damping factor (*d* closer to 1) means the random surfer is less likely to 'teleport' to a random page and is more likely to follow existing links. Thus the PageRank value is more reflective of the actual link popularity within the network.
